```python
#Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Data
data = pd.read_csv("Titanic-Dataset.csv")
```

```python
data.head(7)
```

```python
# Extract the 'Survived' column
survived_counts = data['Survived'].value_counts()


plt.figure(figsize=(4, 3))
survived_counts.plot(kind='bar', color=['red', 'green'], alpha=0.6)

plt.title('Survivals of Titanic Passengers')
plt.xlabel('Survived? ( No= 0 , Yes= 1 )')
plt.ylabel('Number of Passengers')

plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
# Extract the 'Sex' column
survived_counts = data['Sex'].value_counts()


plt.figure(figsize=(4, 3))
survived_counts.plot(kind='bar', color=['#0072BD','red'], alpha=0.7)
plt.title('The gender population of Titanic Passengers')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Filter the data by gender
males = data[data['Sex'] == 'male']
females = data[data['Sex'] == 'female']

# histogram to visualize the age distribution
plt.figure(figsize=(14, 7))

# Male Age Distribution
plt.hist(males['Age'].dropna(), bins=30, edgecolor='black', alpha=0.7, label='Male', color='#0072BD')

# Female Age Distribution
plt.hist(females['Age'].dropna(), bins=30,edgecolor='black', alpha=0.7, label='Female', color='red')

plt.title('Age Distribution of Titanic Passengers by Gender')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(axis='y', alpha=0.6)
plt.show()
```
